```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
KB=0.00008618460742911316
M=11.205e-27 
T=294

T_init=1.499999999999999e+01

```

```python
input = np.loadtxt("energy_and_temperature.dat")
time=input[:,0]
energy=input[:,1]
temperature=input[:,2]
velocity_x = np.loadtxt("velocity_x.dat")
plt.plot(time, energy, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Energy (eV)')
plt.title('Energy fluctuations')
plt.show()
print("deltaE/E = {:e}".format(np.abs(np.std(energy)/np.mean(energy))))
```

```python
plt.plot(time, temperature, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Temperature fluctuations')
plt.show()
print("<T>= ", np.mean(temperature))
```

```python
plt.hist(velocity_x, bins=30, density=True)
x=np.linspace(np.min(velocity_x), np.max(velocity_x), 10000)
plt.plot(x, 1/(np.sqrt(2*np.pi)) * np.sqrt(M/(KB*T))*np.exp(-0.5*M/(KB*T)*x**2), color='red')
plt.title(r'$Q$')
plt.show()
```

```python
def Gamma(vector):
    tm=40
    vectorsquare=np.mean(vector)**2
    Gamma=np.zeros(tm) 
    for i in range(tm):
        for j in range(0,len(vector)-tm):
            Gamma[i]+=vector[j]*vector[i+j]
        Gamma[i]/=(len(vector)-tm)
        Gamma[i]-=vectorsquare

    plt.plot(Gamma/Gamma[0], 'r')

    plt.xlabel(r'$t_M$')
    plt.ylabel(r'$\Gamma(t_M)$')
    plt.title('Funzione di autocorrelazione ')

    plt.show()

Gamma(energy)
Gamma(temperature)

```
