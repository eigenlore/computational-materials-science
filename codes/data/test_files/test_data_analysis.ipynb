```python
import numpy as np
import matplotlib.pyplot as plt
#plt.rcParams['axes.formatter.useoffset'] = False
```

```python
input = np.loadtxt("energy_and_temperature.dat")
time=input[:,0]
energy=input[:,1]
temperature=input[:,2]
velocity_x=input[:,3]
plt.plot(time, energy-np.mean(energy), color='red')
plt.xlabel('Time (s)')
plt.ylabel('Energy (eV)')
plt.title('Energy fluctuations')
plt.show()
print(energy)
```

```python
plt.plot(time, temperature, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Temperature fluctuations')
plt.show()
```

```python
print(np.mean(temperature))
print(np.std(energy)/np.mean(energy))
```

```python
plt.hist(velocity_x, bins=30)
plt.title(r'$Q$')
plt.show()
```

```python
def Gamma(vector):
    tm=40
    vectorsquare=np.mean(vector)**2
    Gamma=np.zeros(tm) 
    for i in range(tm):
        for j in range(0,len(vector)-tm):
            Gamma[i]+=vector[j]*vector[i+j]
        Gamma[i]/=(len(vector)-tm)
        Gamma[i]-=vectorsquare

    plt.plot(Gamma/Gamma[0], 'r')

    plt.xlabel(r'$t_M$')
    plt.ylabel(r'$\Gamma(t_M)$')
    plt.title('Funzione di autocorrelazione ')

    plt.show()

Gamma(energy)
Gamma(temperature)

```
