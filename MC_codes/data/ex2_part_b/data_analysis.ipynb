```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
KB=0.00008618460742911316
```

```python
N=range(2,21)
energy_N=np.zeros(len(N))
chemical_N=np.zeros(len(N))
nn_N=np.zeros(len(N))

for i in range(len(N)):
    input = np.loadtxt("energy_and_nbrsN"+str(N[i])+".dat")
    energy=input[:,0]
    nn=input[:,1]
    energy_N[i]=energy[-1]
    chemical_N[i]=energy_N[i]/N[i]
    nn_N[i]=nn[-1]

    input=np.loadtxt("final_confN"+str(N[i])+".dat")
    x=input[:,0]
    y=input[:,1]
    plt.scatter(x+1, y+1, color='r')
    plt.xlim(0,26)
    plt.ylim(0,26)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.xticks(range(0, 26))
    plt.yticks(range(0, 26))
    plt.title('Final configuration, N='+str(N[i]))
    plt.show()


plt.plot(N, energy_N, color='r')
plt.xlabel('N')
plt.ylabel('Energy (eV)')
plt.title('Energy as a function of N')
plt.show()

plt.plot(N, chemical_N, color='r')
plt.xlabel('N')
plt.ylabel('Chemical potential (eV)')
plt.title('Chemical potential as a function of N')
plt.show()

plt.plot(N, nn_N, color='r')
plt.xlabel('N')
plt.ylabel('Mean number of neighbours')
plt.title('Mean number of neighbours of N')
plt.show()
```
