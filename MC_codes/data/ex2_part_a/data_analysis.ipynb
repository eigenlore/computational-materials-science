```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
KB=0.00008618460742911316
```

```python
input = np.loadtxt("energyN25.dat")
energy=input
plt.plot(energy, color='red')
plt.xlabel('Montecarlo step')
plt.ylabel('Energy (eV)')
plt.title('Energy, N=25')
plt.show()
print('Energy minimum, N=25', np.min(energy))

input=np.loadtxt("init_confN25.dat")
x=input[:,0]
y=input[:,1]
plt.scatter(x+1, y+1, color='r')
plt.xlim(0,26)
plt.ylim(0,26)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Initial configuration, N=25')
plt.grid(True)
plt.xticks(range(0, 26))
plt.yticks(range(0, 26))
plt.show()

input=np.loadtxt("final_confN25.dat")
x=input[:,0]
y=input[:,1]
plt.scatter(x+1, y+1, color='r')
plt.xlim(0,26)
plt.ylim(0,26)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.xticks(range(0, 26))
plt.yticks(range(0, 26))
plt.title('Final configuration, N=25')
plt.show()
```
