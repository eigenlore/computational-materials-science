```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
KB=0.00008618460742911316
N=27
```

```python
J0=['-0.4', '-0.3','-0.2', '-0.1', '-0.0', '0.1', '0.2']
energy_J0=np.zeros(len(J0))
ratio_J0=np.zeros(len(J0))


for i in range(len(J0)):

    input = np.loadtxt("energy_and_nbrsJ0"+J0[i]+"T0.dat")
    energy=input[:,0]
    ratio=input[:,2]/N
    energy_J0[i]=energy[-1]
    ratio_J0[i]=ratio[-1]

    plt.plot(energy, color='r')
    plt.xlabel('MonteCarlo step')
    plt.ylabel('Energy (eV)')
    plt.title('Energy, J0='+str(J0[i]))
    plt.show()


    input=np.loadtxt("final_configJ0"+J0[i]+"T0.dat")
    x=input[:,0]
    y=input[:,1]
    z=input[:,2]
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x+1, y+1, z+1, color='r')
    ax.set_xticks(range(0, 26))
    ax.set_yticks(range(0, 26))
    ax.set_zticks(range(0, 11))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_title('Final configuration, J0='+J0[i]+ ' eV')
    ax.grid(True)
    plt.show()


plt.plot(J0, energy_J0, color='r')
plt.xlabel('J0 (eV)')
plt.ylabel('Energy (eV)')
plt.title('Energy')
plt.show()

plt.plot(J0, ratio_J0, color='r')
plt.xlabel('J0 (eV)')
plt.ylabel('N1/N (eV)')
plt.title('Atom in first layer, T=0K')
plt.show()

```
