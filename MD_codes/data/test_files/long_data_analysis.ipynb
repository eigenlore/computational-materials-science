```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
KB=0.00008618460742911316
M=11.205e-27 
```

```python
input = np.loadtxt("energy_temperaturePBC.dat")
time=input[:,0]
energy=input[:,1]
temperature=input[:,2]
velocity_x = np.loadtxt("velocity_x.dat")
plt.plot(time, energy, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Energy (eV)')
plt.title('Energy fluctuations')
plt.show()
print("deltaE/E = {:e}".format(np.abs(np.std(energy)/np.mean(energy))))
```

```python
plt.plot(time, temperature, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Temperature fluctuations')
plt.show()
print("<T>= ", np.mean(temperature))
```

```python
L=np.loadtxt('angular_momentumPBC.dat')
v_cm=np.loadtxt('v_cmPBC.dat')
for i in range(3):
    print('media v_cm', np.mean(v_cm[:,i]))
    print('err rel v_cm', np.std(v_cm[:,i])/np.mean(v_cm[:,i]))
    plt.plot(time, v_cm[:, i], color='red')
plt.show()


for i in range(3):
    print('media L', np.mean(L[:,i]))
    print('err rel L', np.std(L[:,i])/np.mean(L[:,i]))
    plt.plot(time, L[:, i], color='red')
plt.show()
```
