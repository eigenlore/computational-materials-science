```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
T_INIT=380
DT=4
from matplotlib.animation import FuncAnimation
```

```python
input = np.loadtxt("energy_temperature.dat")
time=input[:,0]
energy=input[:,1]
temperature=input[:,2]
plt.plot(time, energy, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Energy (eV)')
plt.title('Energy fluctuations')
plt.show()
print("deltaE/E = {:e}".format(np.abs(np.std(energy)/np.mean(energy))))
```

```python
plt.plot(time, temperature, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Temperature fluctuations')
plt.show()
print("<T>= ", np.mean(temperature))
```

```python
x_positions=np.loadtxt('x_positions.dat')
y_positions=np.loadtxt('y_positions.dat')
z_positions=np.loadtxt('z_positions.dat')
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(np.min(x_positions), np.max(x_positions))  
ax.set_ylim(np.min(y_positions), np.max(y_positions))  
ax.set_zlim(np.min(z_positions), np.max(z_positions))   
fig.dpi=150

def update(frame):
    ax.clear()
    ax.set_xlim(np.min(x_positions), np.max(x_positions))  
    ax.set_ylim(np.min(y_positions), np.max(y_positions))  
    ax.set_zlim(np.min(z_positions), np.max(z_positions))    
    ax.set_title(f'Time {frame*DT:d} fs')  
    ax.scatter(x_positions[frame], y_positions[frame], z_positions[frame])  

animation = FuncAnimation(fig, update, frames=range(0, len(x_positions), 50),interval=100000)
animation.save('evolution.mp4', writer='ffmpeg', fps=30, dpi=150, codec='libx264', bitrate=8000)
```
