```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
T_INIT = 200
```

```python
DT=[6.25e-17, 1.25e-16, 2.50e-16, 5.00e-16, 1.00e-15, 2.00e-15, 4.00e-15, 8.00e-15]

energy_fluct=np.zeros(len(DT))
mean_temperature=np.zeros(len(DT))

for i in range(len(DT)):
    input=np.loadtxt("energy_temperatureDT"+f"{DT[i]:.2e}"+".dat")
    time=input[:,0]
    energy=input[:,1]
    temperature=input[:,2]
    energy_fluct[i]=np.abs(np.std(energy)/np.mean(energy))
    mean_temperature[i]=np.mean(temperature)
    plt.plot(time, energy, color='red')
    plt.xlabel('Time (s)')
    plt.ylabel('Energy (eV)')
    plt.title('Energy fluctuations, DT= ' + str(DT[i])+ ' s')
    plt.show()
    

```

```python
plt.plot(DT, energy_fluct, color='red')
plt.xlabel('DT (s)')
plt.ylabel(r'$\Delta E/<E>$')
plt.title('Energy fluctuations varying the time step')
plt.show()    
```
