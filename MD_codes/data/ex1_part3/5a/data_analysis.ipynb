```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
KB=0.00008618460742911316
M=11.205e-27 
T_INIT=20
T_INIT=1550
```

```python
input = np.loadtxt("therm_energy_temperatureT20.dat")
time=input[:,0]
energy=input[:,1]
temperature=input[:,2]
plt.plot(time, energy, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Energy (eV)')
plt.title('Energy fluctuations before thermalization')
plt.show()
print("deltaE/E = {:e}".format(np.abs(np.std(energy)/np.mean(energy))))
plt.plot(time, temperature, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Temperature fluctuations before thermalization')
plt.show()
print("<T>= ", np.mean(temperature))
```
