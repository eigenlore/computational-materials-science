```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
fcc=np.loadtxt('../../input_files/fcc100a256.dat')
first_layer = fcc[fcc[:, -1] == np.max(fcc[:, -1])]
```

```python
input = np.loadtxt("energy_temperatureT2000.dat")
time=input[:,0]
energy=input[:,1]
temperature=input[:,2]
plt.plot(time, energy, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Energy (eV)')
plt.title('Energy fluctuations')
plt.show()
print("deltaE/E = {:e}".format(np.abs(np.std(energy)/np.mean(energy))))
plt.plot(time, temperature, color='red')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Temperature fluctuations')
plt.show()
print("<T>= ", np.mean(temperature))
```
