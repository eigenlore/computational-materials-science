```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.formatter.useoffset'] = False
KB=0.00008618460742911316
M=11.205e-27 
T_INIT=15
```

```python
fluct=[]
atoms=[108, 256, 500, 864, 1372 ,2048]
for N in atoms:
    print(f'N={N}')
    input = np.loadtxt("energy_temperatureN"+str(N)+".dat")
    time=input[:,0]
    energy=input[:,1]
    temperature=input[:,2]
    print(np.mean(temperature))
    fluct.append(np.std(temperature)/np.mean(temperature))
```

```python
plt.plot(atoms, fluct, c='r')
plt.title("Temperature fluctuations")
plt.xlabel("Atoms")
plt.ylabel(r"$\frac{\delta T}{T}$")
plt.show()
```
